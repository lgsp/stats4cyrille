# TP Entreposage et fouille de données (STA211)  $ ~~~~~~~~~~~ $ Réseaux de neurones et apprentissage profond

Cours : [http://cedric.cnam.fr/~thomen/cours/STA211.html](http://cedric.cnam.fr/~thomen/cours/STA211.html)

**Les travaux pratiques sont structurés de la manière suivante :**

- La première partie consiste en une introduction à la librairie `Keras` pour entraîner des réseaux de neurones. On abordera des modèles de niveaux croissants de complexité, depuis des réseaux très simples non profonds comme la régression logistique jusqu’à des réseaux convolutifs plus évolués.  
- La seconde partie vise à observer l’effet de « manifold untangling » permis par les réseaux de neurones profonds, qu’on observera en visualisant la distribution des classes dans l’espace d’entrée et dans l’espace des représentations internes des réseaux de neurones.
